```python
# например интересно посмотреть по каким заявителям работает ТГC в Всеволожском р-не по объектам менее 0,1км

pd.read_sql(
    """
    SELECT * FROM appl_km
    LEFT JOIN executors ON appl_km.appl = executors.appl_id
    LEFT JOIN appl_village ON appl_village.appl_village = appl_km.appl
    LEFT JOIN region_settlement ON region_settlement.settlement = appl_village.settlement
    WHERE executor = 'ТГС' AND region = 'Волосовский' AND km < 0.1
    """, 
    sql)
```
