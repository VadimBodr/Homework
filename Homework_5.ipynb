# Задание 3.1 (1 балл)

В последующих заданиях будут учитываться данные начиная с 01.01.1950. Для этого создайте новый `DataFrame`, в котором удалены все строки до 01.01.1950. Используйте созданный DataFrame в последующих заданиях.  

```python
df["Day"] = pd.to_datetime(df["dt"])
df["year"] = df.Day.dt.year
df["month"] = df.Day.dt.month
df["day"] = df.Day.dt.day
df.head()
```

```python
mask = df['year'] >= 1950
df = df[mask]
df

```

# Задание 3.4 (4 балла)

На одной оси постройте графики, отображающие:
1. Зависимость выборочной дисперсии температур от широты города (Latitude).
2. Зависимость средней температуры в январе 1975 года от широты города (Latitude).
3. Зависимость средней температуры в июле 1975 года от широты города (Latitude).

```python
df

```

```python
df.groupby(['Latitude'])['AverageTemperature'].var()

df2 = df[(df['month'] == 1) & (df['year'] == 1975)]
df2 = df2.groupby(['Latitude'])['AverageTemperature'].mean()

df3 = df[(df['month'] == 7) & (df['year'] == 1975)]
df3 = df3.groupby(['Latitude'])['AverageTemperature'].mean()
```

```python
import matplotlib.pyplot as plt
df1 = df.groupby(['Latitude'])['AverageTemperature'].var()
x = df1.index
y1 = df1.values
y2 = df2.values
y3 = df3.values

plt.figure(figsize=(35, 12))
plt.plot(x, y1, color='black', label='Дисперсия')
plt.plot(x, y2, color='red', label='t январь 1975')
plt.plot(x, y3, color='green', label='t июль 1975')
plt.legend()
```
